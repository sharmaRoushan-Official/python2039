### 5. if Statement(Single Condition)


if condition:

    # True Block

```python
age = 18

if age>18:
    print("You are elegible to vote.")
```

```python
20>18
```

```python
age = 18

if age>=18:
    print("You are elegible to vote.")
else:
    print("You are not eleigble to vote")
```

Note:

* if condition is True, code runs

* If False, nothing happens

* Else block runs directly when conditions is False.

* Either if block or Else will be execute --- never both

### 7. if-elif-else (Multiple Conditions)


syntax:
-------

if condition1:
    
    statement1

elif condition2:
    
    statement2

elif condition3:
    
    statment3

else:
    statement4

```python
marks = 95

if marks>=90:
    print("Grade A")
elif marks>= 75:
    print("Grade B")
elif marks>= 60:
    print("Grade C")
else:
    print("Fail")
```

Flow Rule:

* Python check top to bottom
* First True condition Execute
* Remaining Condition are skipped.

### 8. Nested if (condition inside another condition)

syntax:

if codition:

    if condition:
    
        statement

```python
age = 15

citizen = True

if age>=18:   # True
    # print("Elegible to vote")   # True Block 
    if citizen: # True
        print("Elegible to Vote")
    else:
        print("You are not citizen of India")
else:
    print("Not Elegible to vote.")
```

### 9. Short Hand if (One-line condtion)

statement if codition

```python
age = 15


"Elegible to Vote" if age>=18 else "Not Elegible to vote"
```
