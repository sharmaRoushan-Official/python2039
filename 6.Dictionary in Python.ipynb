## 6. Dictionary in Python

### What is dictionary in Python??

A `dictionary` is a collection of key-value pairs.

* Each key is unique.
* Each key maps to a value.
* Dictionary are mutable(unhashable).
* Dictionary are written in curly braces {}.

```python
x = {}
print(x,type(x))
```

```python
s = {1}
print(s,type(s))
```

```python
s = set()
print(s,type(s))
```

```python
student = {
    "name":"Satyadev",
    "age":55,
    "course":"Python"
}
print(student)
```

## 11. Looping Through Dictioanry

For ---> Range complete -- 1-10
While  --->Range not Know --->

```python
student = {
    "name":"Satyadev",
    'age':55,
    "city":"Pune",
    "address":"Kharadi",
    "PIN":411015
}
student
```

```python
for i in student:
    print(i)
```

```python
for key in student.keys():
    print(key)
```

```python
for value in student.values():
    print(value)
```

```python
for key,value in student.items():
    print(key,":",value)
```
