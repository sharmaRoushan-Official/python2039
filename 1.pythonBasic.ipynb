```python
# Operators 

1. Assignment Operator : [=]  # Equals
2. Arithematic Operator: [+,-,*,/,//,**,%]
    1. Unary Operator: [++,--]
    2. Binary Operator: [+,-,*,/,//,**,%]
3. Relational Operator/Comparison Operator: [<,>,<=,>=,!=,==] # Equals to :: True/False
4. Logical Operator: [AND, OR, NOT]  ## True/False
    Univeral Operator: [NAND, NOR]
    Combinational Operator: [XNOR, XOR]
5. Identity Operator: [is , is not]  # Address of variables :: True/False
6. Membership Operator: [in , not in]  # True/False
```

```python
x = 10
x++  ## post increment value 10+1 = 11
x--  ### post Decrement value 10-1 = 9

++x ### pre increment value 1+10 = 11
--x  ## pre Decrement value 1-10 = 9
```
