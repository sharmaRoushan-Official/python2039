Prcedence Rule:
---------------

* Highest Precedence:()
* Medium Precedence: *,/,//,**
* Lowest Precedence: +,-

```python
result = (10+2)*5
# result = 12*5
print(result)
```

```python
result = 10+(2*5)
# result = 10+10
print(result)
```
