* Lists are one of the most used data types in Python

### 8. List methods: 

----

|Method|Work|
|----|----|
|append(x)|Add item at end|
|extend(iterable)|Add multiple items|
|insert(i,x)|Insert at specific index|
|remove(x)|Remove first occurrence value|
|pop(x)|Remove item at x index|
|pop()| Remove last index value|
|clear()| Remove all items at a time.|
|del()| Delete item|
|index(x)| Return index value of x element|
|sort()|Arrange item in ascending or descending order|
|reverse()| Revesrse list|
|copy()|Return shallow copy|

### 9. index() and count()

```python
nums = [10,20,30,40,50,50]

print(nums.index(40))
print(nums.index(50))
print(nums.count(10))
print(nums.count(50))
```
