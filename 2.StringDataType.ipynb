## 2. Python String

### 1. What is a string?

A string is a sequence of characters enclosed in quoatation marks.

A string can contain:

* Letters(a-z,A-Z)
* Numbers(0-9)
* Symbols(@,#,$,₹,%,etc)
* Spaces
* Unicode Characters: utf-8

```python
name = "Ram"
city = "NOIDA"
college = "MIT"

print(name)
print(city)
print(college)
```

### 2. Definition

A string is an immutable(hashable/notchangeable) sequence of Unicode characters.

Important Words:

* Immutable: Cannot be change or modify after creating.
* Sequece -> Characters are stored in order.
* Unicode -> Supports all international langugages.

```python
name = "Python"
print(name)
```

## Indexing in String

> Positive Indexing: [0,1,2,3,4,5,6,7]

> Negative Indexing: [-1,-2,-3,-4,-5]

### String Indexing or Ordered

```python
s = "Python"
print(s)
```

```python
# Positive Indexing

print(s[0])
print(s[1])
print(s[2])
print(s[3])
print(s[4])
print(s[5])
# print(s[6]) Error
```

### Negative Indexing: [-1,-2,-3,-4]

```python
s = "Python"

print(s[-1])
print(s[-2])
print(s[-3])
print(s[-4])
print(s[-5])
print(s[-6])
```
