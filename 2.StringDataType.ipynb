## 2. Python String

### 1. What is a string?

A string is a sequence of characters enclosed in quoatation marks.

A string can contain:

* Letters(a-z,A-Z)
* Numbers(0-9)
* Symbols(@,#,$,₹,%,etc)
* Spaces
* Unicode Characters: utf-8

```python
name = "Ram"
city = "NOIDA"
college = "MIT"

print(name)
print(city)
print(college)
```

### 2. Definition

A string is an immutable(hashable/notchangeable) sequence of Unicode characters.

Important Words:

* Immutable: Cannot be change or modify after creating.
* Sequece -> Characters are stored in order.
* Unicode -> Supports all international langugages.

```python
name = "Python"
print(name)
```

### String Methods

* lower()
* upper()
* capitalize()
* title()
* casefold()

```python
str1 = "hellO WoRLd"
print(str1)


# lower()

strLower = str1.lower()
print(strLower)


# upper()

strUpper = str1.upper()
print(strUpper)

# Capitalize()

strCapitalize = str1.capitalize()
print(strCapitalize)

strTitle = str1.title()
print(strTitle)

# casefold()
strCasefold = str1.casefold()
print(strCasefold)
```

### 20. Checking Methods

* isalpha()
* isdigit()
* islower()
* isnum()
* isspace()
* istitle()
* islower()
* isupper()
* startswith()
* endswith()

```python
a = "abc" # lower
b = "Abc D" # upper
c = "123" # digit
d = "abc123ABC" # alpha numeric
e = " "  # space

print(a.isalpha())
print(a.islower())
print(a.isupper())

print(b.isupper())
print(c.isdigit())
print(a.isalnum())  # if alphabet or number 
print(a.isalnum())
print(d.isalnum())
print(e.isspace())
print(a.istitle())
print(b.istitle())
```

```python
a = "Apple"
b = "banana"
# startswith()
# endswith()

print(a.startswith("A"))
print(a.startswith("B"))  # False
print(b.startswith("B"))
print(b.startswith("b"))

print(a.endswith("e"))
print(b.endswith('a'))
```
