## 4. Tuples in Python


#### 1. What is Tuple?

A `tuple` is a `collection data type in python` used to sotre multiple values in a single variable.


### Features

* Ordered Collection

* Immutable(hashable/Not changeable): Cannot be change after creation

* Allows duplicate values

* Can store different data types

* Faster than lists

* Written using round brackets.():parentheses

```python
numbers = ()
print(numbers)
```

```python
print(type(numbers))
```

### 2. Creating Tuples

##### Method-1: Using Parentheses



```python
fruits = ("apple","mango","banana")
print(fruits)
```

```python
print(type(fruits))
```

### Method- 2- Without Parentheses

Parentheses are optional

```python
colors = "Red","green","Blue"
print(colors)
print(type(colors))
```

### Method-3 : tuple() constructor

Tuples can store multiple data types

```python
data = ("John",25,75.5,True)
print(data)
```

```python
x = ["John",25,75.5,True]
print(x)
```

```python
tup1 = tuple(x)
print(tup1)
```

#### 4. Nested Tuples

A tuple inside another tuple

```python
nested_tuple = (1,2,(3,4),(5,6))
print(nested_tuple)
```

#### 5. Accessing Tuple Elements

###### Positive and Negative

```python
fruits = ("apple","mango","banana",10)
print(fruits)


print(fruits[0])
print(fruits[1])
print(fruits[2])
```

```python
print(type(fruits[0]))
print(type(fruits[1]))
print(type(fruits[2]))
print(type(fruits[3]))
```

Negative Indexing



```python
print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
print(fruits[-4])
```

### 6. Tuple Slicing 
Used to extract a protion of the tuple


syntax

tuple[start:stop:step]

```python
numbers = (10,20,30,40,50,60)

print(numbers,type(numbers))
```

```python
print(numbers[3:5])
```

### 7. Tuple Immutability

Tuples cannot be modified after creating

```python
numbers = (10,20,30)

numbers[0] = 100
print(numbers)
```

```python
numbers = [10,20,30]

numbers[0] = 100
print(numbers)
```

### 8. Changing tuple Elements 



```python
numbers = (10,20,30)
print(numbers)
print(type(numbers))


# list() constructor

temp = list(numbers)
print(temp,type(temp))

temp[0] = "Swadhin"
print(temp)

# tuple() constructor

number2 = tuple(temp)
print(number2)
```

### 9. Looping Through Tuples

Using for loop

```python
fruits = ('apple',"banana","mango")
print(fruits)
```

```python
for fruit in fruits:
    print(fruit)
```

```python
for fruit in range(len(fruits)):
    print(fruit, fruits[fruit])
```

### 10. Tuple Length

uses len()

```python
numbers = (10,20,30,40,50,60,70,80,90)
print(numbers)
```

```python
len(numbers)
```

### 11.Packing 

Packing means storing multiple values into a tuple

```python
data = ('Ankesh',25,"Python Developer")
print(data)
print(type(data))
```

### 12. Tuple Unpacking 
Exract tuple values into variables.

```python
data = ('Ankesh',25,"Python Developer")
print(data)

# unpacking
name,age,job = data
print(name)
print(age)
print(job)
```

### 13. Extended Unpacking

Using * Operator (Asterisk)

```python
numbers = (1,2,3,4,5)
print(numbers)

a,*b,c = numbers  # *b --- list
print(a)
print(b)
print(c)
```

```python
numbers = (1,2,3,4,5)
print(numbers)

a,b,*c = numbers  # *b --- list
print(a)
print(b)
print(c)
```

```python
numbers = (1,2,3,4,5)
print(numbers)

*a,b,c = numbers  # *b --- list
print(a)
print(b)
print(c)
```

## 14. Tuple Concatenation

Combine tuples using +



```python
t1 = (1,2,3)
t2 = (4,5,6)

result = t1+t2
print(result)
```

## 15. Tuple Repetition

Using * operator

```python
numbers = (1,2,3)

print(numbers*3)
```

### 15. Membership Operator- [in or not in]


```python
fruits = ('apple',"banana","mango")
print(fruits)
```

```python
print("Banana" in fruits)
```

```python
print("banana" not in fruits)
```

```python
print("banana" in fruits)
```

### 17. Tuple Methods

Tuple have only two built-in methods

1. count()

2. index()

```python
numbers = (1,2,3,4,5,2,2,2)

print(numbers)
```

```python
print(numbers.count(2))  # total number occurrence in the tuple
```

```python
# index() 

print(numbers.index(2))
```

```python
print("Hello world")  # Built-in Function
```

```python
numbers.index(2) # index - method 
```

```python
tup1 = (10,20,30,40)
sum(tup1)
```

```python
tup1.count(20)
```

### 18. Built-in Functions with Tuples

* max()
* min()
* sum()
* sorted()
* len()

```python
numbers = (10,20,30)
print(numbers)
```

```python
print(max(numbers))
print(min(numbers))
print(len(numbers))
print(sum(numbers))
```

```python
numbers = (10,20,30,5)
print(numbers)

print(sorted(numbers)) # default-Ascending order 
```

### 20. Advantage of tuples

* Faster than lists
* Immutable (safe from accidental change)
* Can be used as dictionary key. [because keys are immutable]
* Consume less memory


### 21. Tuple vs List

|Feature|Tuple|List|
|-----|------|----|
|Brackets|()|[]|
|Mutable|No|Yes|
|Speed|Faster|Slower|
|Memory|Less|More|


## 22. When Use Tuples

Use tuples when:

* Data should not change.
* Fixed Records(Coordinates)
* Returning multiple values from functions

```python
import timeit
```

```python
print(timeit.timeit("(1,2,3,4,5)",number=100000))
print(timeit.timeit("[1,2,3,4,5]",number=100000))
```
