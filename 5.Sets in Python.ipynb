### 5. Sets in Python

### 1. Introduction of Sets

A `set` in Python is a collection data type used to store multiple unique elements.

It is based on the mathematical concept of a set.


##### Key Features:

1. Unordered
2. Unidexed
3. Mutable(Unhashable/changeable) --> can add/remove items
4. Does not allow duplicate value.
5. Elements must be immutable(hashable)


Example

```python
s = {10,20,30}
print(s)
print(type(s))
```

#### 2. Creating Sets

Method 1: Using Curly Bracket- {}

```python
s = {1,2,3,4,5,6,7,8,9}
print(s)
print(type(s))
```

Method 2: set() constructor

```python
list1 = [10,10,10,20,30,70,80,80,90]
print(list1,type(list1))


s = set(list1)  # list is converted into set
print(s,type(s))
```

Important

Empty set cannot be crated using {}

```python
s = {}
print(s,type(s))
```

```python
s = {1}
print(s,type(s))
```

```python
# set() constructor


x = set()
print(x,type(x))
```

### 10. Set Operations

Python sets support mathematical Operations

1. Union
2. Intersection
3. difference
4. symmetric_difference

```python
s1 = {1,2,3}
s2 = {3,4,5}

# method() and symbol
# .union() and |


s1.union(s2)
```

```python
# symbol

s1|s2  # | ---> Union of two sets 
```

```python
s1 = {1,2,3}
s2 = {3,4,5}

# intersection return common elements
# method() and symbol()
# .intersection() and &


s1.intersection(s2)
```

```python
# symbol

s1&s2
```

```python
s1 = {1,2,3}
s2 = {3,4,5}

# a1-b1

# a-b
# b-a

# method() and symbol
# .difference() and -
print(s1.difference(s2))
print(s2.difference(s1))
```

```python
# symbol

s1-s2
```

```python
s2-s1
```

```python
## symmetric Difference
# method() and symbol()
# .symmetric_difference and ^ -- carot
```

```python
s1 = {1,2,3}
s2 = {3,4,5}

s1.symmetric_difference(s2)
```

```python
# symbol 

s1^s2
```

### 11. Set Comparison Operations 

subset and superset

```python
A = {1,2}
B = {1,2,3}

print(A.issubset(B))
```

```python
print(B.issubset(A))
```

```python
print(B.issuperset(A))
```

```python
print(A.issuperset(B))
```

## disjoint

Checks if no common elements

```python
A = {1,2}
B = {3,4}

print(A.isdisjoint(B))
```

```python
A = {1,2,3}
B = {3,4}

print(A.isdisjoint(B))
```

### 12. Frozen Set()

A frozen set is an immutable version of set.

Example:

```python
fs = frozenset([10,10,20,30,40])
fs
```

```python
fs.add(20)
```

```python
fs.remove(20)
```

## 17. when to Use Sets

Use Sets when:
