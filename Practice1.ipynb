```python
### 10. Write a Python program to check if a person is a *child (≤12), **teenager (13–19), or **adult (≥20)*.

age = int(input("Enter your age: "))

if age>=20:
    print(f"You are adult:{age}")
elif age>=13 and age<=19:
    print(f"You are teenager:{age}")
elif age>0 and age<12:
    print(f"You are child:{age}")
else:
    print(f"Enter a valid age:{age}")
```

```python
## 11. Write a Python program to find the *largest among three numbers*.

num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))


if num1>num2 and num1>num3:
    print(f"{num1} is greater than {num2} and {num3}")
elif num2>num1 and num2>num3:
    print(f"{num2} is greater than {num1} and {num3}")
elif num3>num1 and num3>num2:
    print(f"{num3} is greater than {num1} and {num2}")
```

```python
##12. Write a Python program to check whether a number is *positive and even, **positive and odd, or **negative*.

# 10 -- positive and even

# -20 -- negative and even

num1 = int(input("Enter any number: "))

if num1>=0:
    if num1%2==0:
        print(f"{num1} is even number and Positive")
    else:
        print(f"{num1} is odd number and Positive")
else:
    if num1%2==0:
        print(f"{num1} is even number and Negative")
    else:
        print(f"{num1} is odd number and Negative")
```
